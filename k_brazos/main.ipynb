## Introducción al Problema

El problema del bandido multibrazo (k-armed bandit) modela una situación de toma de decisiones secuencial bajo incertidumbre, constituyendo el ejemplo fundamental del dilema entre **Exploración** y **Explotación** en el Aprendizaje por Refuerzo.

En cada paso temporal $t$, el agente debe seleccionar una acción (brazo) $a_t \in \{1, ..., k\}$. Al hacerlo, recibe una recompensa $r_t$ extraída de una distribución de probabilidad desconocida $\mathcal{P}_{a_t}$ asociada a ese brazo específico.El objetivo del agente es maximizar la recompensa acumulada a lo largo de un horizonte temporal $T$, o equivalentemente, minimizar el *regret* (rechazo), que cuantifica la pérdida por no haber seleccionado siempre el brazo óptimo desde el principio.

En esta sección se evalúa el rendimiento empírico de diversas familias algorítmicas (Greedy, UCB y basadas en gradiente/probabilidad) frente a entornos estocásticos con diferentes topologías de recompensa.
