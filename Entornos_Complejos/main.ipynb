## Introducción
En esta fase del proyecto, vamos más allá del problema del bandido de múltiples brazos para abordar entornos MDP (procesos de decisión de Markov) completos. El objetivo es analizar cómo los agentes aprenden a navegar en entornos espaciales y cómo escalan hacia problemas de alta complejidad.

El trabajo se divide en dos estudios principales:


### 1. [Métodos tabulares: Entorno en SimpleGrid](metodos_tabulares.ipynb)
En este estudio utilizamos el entorno `gym-simplegrid` para evaluar algoritmos clásicos.
* **Algoritmos:** Q-Learning, SARSA, Monte Carlo (On-policy y Off-policy).
* **Enfoque:** Comparación de las trayectorias óptimas y la velocidad de convergencia hacia la política objetivo.

### 2. [Métodos aproximados: Entorno en Flappy Bird](metodos_aproximados.ipynb)
Experimentación con la aproximación funcional para gestionar espacios de estado continuos o de alta dimensionalidad.
* **Algoritmo:** Deep Q-Network (DQN) y Sarsa Semi-gradient.
* **Enfoque:** Estudio del hiperparámetro *Experience Replay* (Memory Size) y estabilidad del aprendizaje en un `Flappy Bird`.
